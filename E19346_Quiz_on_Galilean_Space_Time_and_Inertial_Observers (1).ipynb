###**Question No.4**

```python
import plotly.graph_objects as go

origin = [0, 0, 0]

x_axis = [2, 0, 0]
y_axis = [0, 2, 0]
z_axis = [0, 0, 2]

trace_x = go.Scatter3d(x=[origin[0], x_axis[0]], y=[origin[1], x_axis[1]], z=[origin[2], x_axis[2]], mode='lines', name='X-axis')
trace_y = go.Scatter3d(x=[origin[0], y_axis[0]], y=[origin[1], y_axis[1]], z=[origin[2], y_axis[2]], mode='lines', name='Y-axis')
trace_z = go.Scatter3d(x=[origin[0], z_axis[0]], y=[origin[1], z_axis[1]], z=[origin[2], z_axis[2]], mode='lines', name='Z-axis')

fig = go.Figure(data=[trace_x, trace_y, trace_z])
fig.show()
```

###**Question No.5**

```python
import numpy as np
import plotly.graph_objects as go

#Velocity vector of e prime frame
v=np.array([0,0,1])

w=5;

#Time Values
tt=np.linspace(0,4,100)

zeta=np.zeros((len(tt),3));

for i in range (len(tt)):
    #with respect to e
    a=np.array([np.sin(w*tt[i]),np.cos(w*tt[i]),0]);

    #with respect to e prime
    zeta[i]=(a-v*tt[i]);


x_vals = np.zeros(len(tt))
y_vals = np.zeros(len(tt))
z_vals = np.zeros(len(tt))

for i in range(len(tt)):
  x_vals[i] = zeta[i][0]
  y_vals[i] = zeta[i][1]
  z_vals[i] = zeta[i][2]

axis_range = [[-3,3],[-3,3],[-3,3]]
fig_title = 'Motion of the particle in e-prime frame'

trace_particle = go.Scatter3d(
    x=[x_vals[0]], y=[y_vals[0]], z=[z_vals[0]],
    mode="markers",
    marker=dict(color="red", size=10)
)

trace_path = go.Scatter3d(
    x=x_vals, y=y_vals, z=z_vals,
    mode="lines",
    line=dict(color="blue", width=2),
    name='Path'
)

layout = go.Layout(
    title_text=fig_title,
    hovermode="closest",
    updatemenus=[dict(
        type="buttons",
        buttons=[dict(
            label="Play",
            method="animate",
            args=[None]
        )]
    )],
    scene=dict(
        xaxis=dict(range=axis_range[0], autorange=False),
        yaxis=dict(range=axis_range[1], autorange=False),
        zaxis=dict(range=axis_range[2], autorange=False),
        aspectratio=dict(x=1, y=1, z=1)
    )
)

frames = [go.Frame(
    data=[go.Scatter3d(
        x=[point[0]], y=[point[1]], z=[point[2]],
        mode="markers",
        marker=dict(color="red", size=10),
        name='Particle'
    )]) for point in zeta]

fig = go.Figure(data=[trace_particle, trace_path], layout=layout, frames=frames)
fig.show()

```
